Dilasi Tanpa Library

```python
def dilasi_citra(F,w):
  #size image
  p,q = F.shape

  imgD= np.zeros((p,q), dtype=np.uint8)

  #Generate structure element dilasi (SED)
  SED= np.ones((w,w), dtype=np.uint8)
  constant1= (w-1)//2

  #Proses Dilasi
  for i in range(constant1, p-constant1):
    for j in range(constant1,q-constant1):
      temp= F[i-constant1:i+constant1+1, j-constant1:j+constant1+1]
      product= temp*SED
      imgD[i,j]= np.max(product)
  return imgD

img_d= cv2.imread('/content/drive/MyDrive/UniLife/Semester5/PCVK/assets/Images/platnomer.jpg',0)
ret, thresh = cv2.threshold(img_d, 125, 200, cv2.THRESH_BINARY)
w=5
imgD = dilasi_citra(thresh,w)

plt.subplot(131),plt.imshow(img_d,cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(imgD,cmap = 'gray')
plt.title('Citra Hasil Dilasi'), plt.xticks([]), plt.yticks([])
plt.show()
```

Erosi Tanpa Library

```python
def erosi_citra(F,k):
  #size image
  m,n= F.shape

  #Generate structure element erosi (SE)
  SE= np.ones((k,k), dtype=np.uint8)
  constant= (k-1)//2

  imgE= np.zeros((m,n), dtype=np.uint8)

  #Proses Erosi
  for i in range(constant, m-constant):
    for j in range(constant,n-constant):
      temp= F[i-constant:i+constant+1, j-constant:j+constant+1]
      product= temp*SE
      imgE[i,j]= np.min(product)
  return imgE

img_e= cv2.imread('/content/drive/MyDrive/UniLife/Semester5/PCVK/assets/Images/platnomer.jpg',0) #
ret1, thresh1 = cv2.threshold(img_e, 175, 225, cv2.THRESH_BINARY)
k=5
imgE = erosi_citra(thresh1,k)

plt.subplot(131),plt.imshow(img_e,cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(imgE,cmap = 'gray')
plt.title('Citra Hasil Erosi'), plt.xticks([]),
plt.yticks([])
plt.show()
```

```python
import cv2 as cv
from matplotlib import pyplot as plt
from skimage import filters
```

```python
from skimage.morphology import skeletonize
# read image
img = cv.imread('/content/drive/MyDrive/UniLife/Semester5/PCVK/assets/Images/lily.jpg',0)

binary = img > filters.threshold_triangle(img)

# true false to one
binary_cp = binary.copy()
binary_cp[binary_cp == True] = 1
binary_cp[binary_cp == False] = 0

# skeletonize image
skeleton = skeletonize(binary_cp)

# print images
plt.figure(figsize=(20,20))

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(122),plt.imshow(skeleton,cmap = 'gray')
plt.title('skeleton Image'), plt.xticks([]), plt.yticks([])

plt.savefig("binary and bin_skeleton.png")
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('/content/drive/MyDrive/UniLife/Semester5/PCVK/images/KTP7.jpeg', 0)

# Thresholding
_, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(img, cmap='gray')
plt.title("Citra Asli KTP")
plt.show()

plt.imshow(binary_img, cmap='gray')
plt.title("Citra Biner KTP")
plt.show()
```

Dilasi

```python
def dilasi_citra(F, w):
    p, q = F.shape
    imgD = np.zeros((p, q), dtype=np.uint8)
    SED = np.ones((w, w), dtype=np.uint8)
    constant1 = (w - 1) // 2

    for i in range(constant1, p - constant1):
        for j in range(constant1, q - constant1):
            temp = F[i - constant1:i + constant1 + 1, j - constant1:j + constant1 + 1]
            imgD[i, j] = np.max(temp * SED)
    return imgD

# Dilasi dengan ukuran SED 3x3
dilated_img = dilasi_citra(binary_img, 3)
plt.imshow(dilated_img, cmap='gray')
plt.title("Citra Hasil Dilasi")
plt.show()
```

Erosi

```python
def erosi_citra(F, k):
    m, n = F.shape
    SE = np.ones((k, k), dtype=np.uint8)
    constant = (k - 1) // 2
    imgE = np.zeros((m, n), dtype=np.uint8)

    for i in range(constant, m - constant):
        for j in range(constant, n - constant):
            temp = F[i - constant:i + constant + 1, j - constant:j + constant + 1]
            imgE[i, j] = np.min(temp * SE)
    return imgE

# Erosi dengan ukuran SE 3x3
eroded_img = erosi_citra(binary_img, 3)
plt.imshow(eroded_img, cmap='gray')
plt.title("Citra Hasil Erosi")
plt.show()
```

```python
# Opening: erosi lalu dilasi
opened_img = dilasi_citra(erosi_citra(binary_img, 3), 3)
plt.imshow(opened_img, cmap='gray')
plt.title("Citra Hasil Opening")
plt.show()

# Closing: dilasi lalu erosi
closed_img = erosi_citra(dilasi_citra(binary_img, 3), 3)
plt.imshow(closed_img, cmap='gray')
plt.title("Citra Hasil Closing")
plt.show()
```
